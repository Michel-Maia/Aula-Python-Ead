```python
df_titanic = pd.read_csv("titanic.csv") 

print(df_titanic.info())
df_titanic.head()
```

 removendo possÃ­veis linhas duplicadas na base de dados

```python
df_titanic.drop_duplicates(keep='last', inplace=True)
```

transformar todos os gÃªneros em maiÃºsculas e guardar dentro da prÃ³pria coluna (sobrescrevendo a coluna), e tambÃ©m vamos converter todas as idades para inteiro.

```python
df_titanic['Sex'] = df_titanic['Sex'].str.upper()

df_titanic['Age'] = df_titanic['Age'].astype(int)

print(df_titanic.dtypes)
df_titanic.head(2)
```

representar a presenÃ§a de parentes (independente se irmÃ£o, cÃ´njuge ou filhos) em uma Ãºnica coluna. Vamos entÃ£o criar a coluna "n_relatives"

```python
df_titanic['n_relatives'] = df_titanic['Siblings/Spouses Aboard'] + df_titanic['Parents/Children Aboard']

df_titanic.head()
```

nova coluna classificando a idade do passageiro. Dependendo da idade, ele serÃ¡ uma crianÃ§a ou adolescente ou adulto ou idoso. Se o passageiro tiver atÃ© 10 anos, ele deve ser classificado como crianÃ§a; de 10 a 18, como adolescente; de 18 a 65, como adulto; e acima de 65, idoso.

A biblioteca pandas possui o mÃ©todo cut(), que "corta" os valores classificando de acordo com os parÃ¢metros.

A construÃ§Ã£o do mÃ©todo Ã© feita por pandas.cut(x, bins, right: bool = True, labels=None, retbins: bool = False, precision: int = 3, include_lowest: bool = False, duplicates: str = 'raise'). 

```python
idade_min = df_titanic['Age'].min()
idade_max = df_titanic['Age'].max()

intervalos = [idade_min-1, 10, 18, 65, idade_max]
rotulos = ['crianÃ§a', 'adolescente', 'adulto', 'idoso']

df_titanic['faixa_idade'] = pd.cut(x=df_titanic['Age'], bins=intervalos, labels=rotulos)

print(df_titanic.info())
df_titanic.head()
```

Ao utilizar o mÃ©todo cut, criamos uma variÃ¡vel categÃ³rica. Uma variÃ¡vel categÃ³rica assume um nÃºmero limitado e geralmente fixo de valores possÃ­veis, por exemplo, gÃªnero, classe social, tipo sanguÃ­neo, afiliaÃ§Ã£o a paÃ­s, faixa etÃ¡ria, etc. A biblioteca pandas possui um tipo de dados chamado Categorical. Lembre-se de que tipos de dados sÃ£o contÃªineres e que mÃ©todos podem ser polifÃ³rmicos (diferentes comportamentos para diferentes objetos).

Para saber a quantidade de passageiros que sobreviveram e a quantidade que nÃ£o teve o mesmo destino, vamos criar dois filtros, utilizando a coluna Survived. Com os filtros, selecionamos os registros de cada grupo. EntÃ£o podemos usar o atributo shape, que retorna uma tupla com a quantidade de linhas e colunas, e selecionar somente as linhas (shape[0])

```python
# Qual a quantidade de sobreviventes e nÃ£o sobreviventes?

filtro_sobreviventes = df_titanic['Survived'] == 1
filtro_nao_sobreviventes = df_titanic['Survived'] == 0

qtde_sobreviventes = df_titanic.loc[filtro_sobreviventes].shape[0]
qtde_nao_sobreviventes = df_titanic.loc[filtro_nao_sobreviventes].shape[0]

print('Quantidade de sobreviventes = ', qtde_sobreviventes)
print('Quantidade de nÃ£o sobreviventes = ', qtde_nao_sobreviventes)
```

```python
# Quantos idosos estavam a bordo e quantos sobreviveram?

filtro_sobreviventes = df_titanic['Survived'] == 1
filtro_idoso = df_titanic['faixa_idade'] == 'idoso'

qtde_idosos = df_titanic.loc[filtro_idoso].shape[0]
qtde_idosos_sobreviventes = df_titanic.loc[(filtro_sobreviventes) & (filtro_idoso)].shape[0]

print('Quantidade de idosos a bordo = ', qtde_idosos)
print('Quantidade de idosos que sobreviveram = ', qtde_idosos_sobreviventes)
```

Para saber, dentre as pessoas que sobreviveram, qual foi a taxa de homens e de mulheres, precisamos extrair as trÃªs informaÃ§Ãµes (quantidade total de sobreviventes, de sobreviventes homens, de mulheres e calcular a porcentagem)

```python
# Das pessoas que sobreviveram, qual a taxa de sobrevivÃªncia entre homens e mulheres?

filtro_sobreviventes = df_titanic['Survived'] == 1
filtro_homem = df_titanic['Sex'] == 'male'
filtro_mulher = df_titanic['Sex'] != 'male'

qtde_homens_sobreviveram = df_titanic.loc[(filtro_sobreviventes) & (filtro_homem)].shape[0]
qtde_mulheres_sobreviveram = df_titanic.loc[(filtro_sobreviventes) & (filtro_mulher)].shape[0]

# Para calcular a taxa de sobreviventes, precisamos do valor total de sobreviventes
qtde_passageiros_sobreviveram = df_titanic[filtro_sobreviventes].shape[0]

taxa_homens = qtde_homens_sobreviveram * 100 / qtde_passageiros_sobreviveram
taxa_mulheres = qtde_mulheres_sobreviveram * 100 / qtde_passageiros_sobreviveram

print("A taxa de sobreviventes homens foi = ", taxa_homens)
print("A taxa de sobreviventes mulheres foi = ", taxa_mulheres)
```

Para descobrir o valor mÃ©dio pago pelas passagens dos que sobreviveram e nÃ£o, podemos filtrar as linhas e selecionar a coluna 'Fare', aplicando o mÃ©todo mean() para extrair a mÃ©dia

```python
# Qual foi o valor mÃ©dio da passagem entre os que sobreviveram e os que nÃ£o sobreviveram?

filtro_sobreviventes = df_titanic['Survived'] == 1
filtro_nao_sobreviventes = df_titanic['Survived'] == 0

valor_medio_sobreviventes = df_titanic.loc[filtro_sobreviventes, 'Fare'].mean()
valor_medio_nao_sobreviventes = df_titanic.loc[filtro_nao_sobreviventes, 'Fare'].mean()

print('Valor mÃ©dio das passagens dos que sobreviveram = ', valor_medio_sobreviventes)
print('Valor mÃ©dio das passagens dos que nÃ£o sobreviveram = ', valor_medio_nao_sobreviventes)
```
